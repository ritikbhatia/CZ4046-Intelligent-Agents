### Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Opening recorded data 

```python
policy_iteration_data = pd.read_csv("../recorded_data/policy_iteration.csv")
```

```python
policy_iteration_data.head()
```

```python
policy_iteration_data.tail()
```

### Plotting relevant data

```python
plt.figure(figsize=(16, 8))
for cell in policy_iteration_data:
    plt.plot(policy_iteration_data[cell])

plt.legend(policy_iteration_data, loc='center left', bbox_to_anchor=(1, 0.5))

plt.title('Utility estimates as a function of number of iterations')
plt.xlabel('Iteration Number')
plt.ylabel('Utility estimate at each cell')

plt.show()
```
